```python
# Source: https://diwashrestha.com/2017/09/18/machine-learning-on-iris/
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris") #Load dataset
print(iris.head()) #Print first few entries of dataset
```

```python
print iris.shape #dimensions of dataset
```

```python
outputs=iris.species; #outputs or features or classes
```

```python
inputs=iris.drop('species', axis=1); #inputs
```

```python
from sklearn.model_selection import train_test_split
in_train, in_test, out_train, out_test = train_test_split(inputs, outputs, 
                                                    test_size=0.6, 
                                                    random_state=100, 
                                                    stratify=outputs) #test_size specification, randomized sampling of outputs (data stratification strategy)
```

```python
sns.set(style="ticks") #Seaborn plots that are really pretty
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species",palette="bright")
plt.show()
```

```python
#More seaborn plots
piris = pd.melt(iris, "species", var_name="measurement") 
sns.factorplot(x="measurement", y="value", hue="species", data=piris, height=7, kind="bar",palette="bright") 
plt.show() 
print(piris.head())
```

```python
clf = tree.DecisionTreeClassifier() #Decision tree classifier
clf.fit(in_train,out_train) #Train decision tree classifier based on inputs -> outputs
```

```python
from sklearn.datasets import load_iris #This may be a redundant step
iris=load_iris()
tree.export_graphviz(clf,
out_file='iris.dot',  feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)
```

```python
out_pred = (clf.predict(in_test)) #Predict outputs from test (or validation) inputs
```

```python
print ('Accuracy Score');
print (accuracy_score(out_test, out_pred)* 100); #Print accuracy score.  
                                                 #The accuracy score is higher if a 
                                                 #larger portion of the original dataset is used in training
```

```python
#Save the prediction package
joblib.dump(clf, 'iris.pkl')
#['iris.pkl']
```

```python
clf2 = joblib.load('iris.pkl') #Load save classification package
clf2.predict(in_test) #Predict 
```
